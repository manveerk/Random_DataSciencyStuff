```python
# DS is an algo that finds best fit line for dataset, or  m & b values with minimum cost
# MSE = 1/n sum(y - (mx+b))**2
#Partial derivatives
#d/dm = 2/n Sum(-xi (yi-yp)) yi is actual y and yp is predicted
#d/db = 2/n Sum -(yi-yp)
# m = m-learning rate * d/dm 
# b = b-learning rate * d/db
```

```python
import numpy as np
```

```python
def grad_desc(x,y):
    m_curr= b_curr = 0
    iterations = 100
    learning_rate = 0.08
    n = len(x) # assumption len of x and y is same
    for i in range(iterations):
        y_predicted = m_curr * x +b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        dm = -(2/n)* sum(x*(y-y_predicted))
        db = -(2/n)* sum(y-y_predicted)
        m_curr = m_curr - learning_rate * dm
        b_curr = b_curr - learning_rate * db
        print('m is {} b is {} cost is {}'.format(m_curr, b_curr,cost))
        
```

```python
x = np.array([1,2,3,4,5])
print(len(x))
y = np.array([5,7,9,11,13])
grad_desc(x,y)
```
